# Scatter Plot

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0,10,30)
y= np.sin(x)
plt.plot(x,y,'o',color='pink')
plt.show()
```

```python
plt.plot(x,y,'-o',color='lightblue')
plt.show()
```

```python
plt.plot(x,y,'p',color='blue')
plt.show()
```

```python
plt.plot(x,y,'-p',color='red')
plt.show()
```

```python
plt.scatter(x,y)
```

```python
data = np.random.randn(1000)
plt.hist(data,bins=100,histtype='bar')
plt.show()
```

```python
x= np.linspace(0,10,1000)
plt.plot(x,np.sin(x),label='Sin')
plt.plot(x,np.cos(x),label='Cos')
plt.legend(loc='lower left',ncol=2,fancybox=True,borderpad=0.5,shadow=True)
plt.show()
```

```python
x= np.linspace(0,10,1000)
inp = input("Enter your Trignometric Function")
if inp == 'Sin':
    plt.plot(x,np.sin(x),label='Sin')
elif inp == 'Cos':
    plt.plot(x,np.cos(x),label='Cos')
elif inp =='Both':
    plt.plot(x,np.sin(x),label='Sin')
    plt.plot(x,np.cos(x),label='Cos')


plt.legend(loc='lower left',ncol=2,fancybox=True,borderpad=0.5,shadow=True)
plt.show()
```
